# ABC Tunes GUI Notebook
Interactive interface.

Initiate the database

```python
import manager
bm = manager.BookManager('abc_books')
bdb = manager.BookDatabase('books.db')
books = bm.search()
tunes = bm.parse(books)
bdb.setup()
bdb.insert_bulk(tunes)
bdb.get_all_tunes()
```

Search for rhythm

```python
search = input("Enter rhythm to search for: ")
bdb.get_tunes_by_rhythm(search)
```

Search by title (case insensitive)

```python
search = input("Enter search term: ")
bdb.search_tunes(search)
```

Search by book number

```python
search = input("Enter book number to search for: ")
bdb.get_tunes_by_book(search)
```
